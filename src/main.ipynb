### psql で SQL 実行結果を CSV に保存する流れ

### 1. psql に接続
Docker の Postgres コンテナに入って `psql` を起動する。

bashで
docker exec -it sql_practice_db psql -U user -d practice
→ practice=# というプロンプトになればOK。
### 2.\copy コマンドの使い方
COPY → サーバー側にファイルを出力（コンテナ内のパス）
\COPY → psql クライアント側から実行、マウントしてあればホストのディレクトリに保存できる

今回は docker-compose.yml で ./export:/export をマウントしてあるので、
保存先を /export/xxx.csv に指定するとホスト側の ./export/ に出力される。

### 3. csvで出すときこれを打てばexportに作られる
\copy (
  WITH sales AS (
    SELECT
      c.country AS country,
      g.name    AS genre,
      il.quantity * il.unitprice AS revenue
    FROM invoiceline il
    JOIN invoice  i ON i.invoiceid  = il.invoiceid
    JOIN customer c ON c.customerid = i.customerid
    JOIN track    t ON t.trackid    = il.trackid
    JOIN genre    g ON g.genreid    = t.genreid
  )
  SELECT country, genre, SUM(revenue) AS revenue
  FROM sales
  GROUP BY country, genre
) TO '/export/country_genre_revenue.csv' WITH CSV HEADER;
